# **y = X.w**

where y is a single columned vector with all outputs in each row

X is a matrix of rows = rows of y. filled with feature values. the first column of X is filled with 1s. so, the no. of columns in X = (No of features + 1)

w is a vector with no of rows = no of columns of X

In linear regression , the task is to find unknowns(w vector)

y = X.w   --> (1)

**[find inverse of X = X^-1]**

multiply (1) with X^-1 on both sides

X^-1 . y = X . X^-1 . w

**[when a matrix is multiplied with its inverse, the result is a identity matrix]**

X^-1 . y = I . w

**[any vector multiplied with a identity matrix would result the *vector itself*]**

X^-1 . y = w

this is how we find the unknowns of a linear system of n no of unknowns **computationally**

## **Affine Transformations via matrix application**

using `plot_vectors()` , we can plot a vector on a graph

```python
def plot_vectors(vector1, color):
  plt.figure()
  plt.axvline(x=0, color='lightgray')
  plt.axhline(y=0, color='lightgray')

  for i in range(len(vector1)):
      x = np.concatenate([[0,0],vector1[i]])
      plt.quiver([x[0]], [x[1]], [x[2]], [x[3]],angles='xy', scale_units='xy', scale=1, color=color[i],)
```

```python
v  = np.array([1,3]) # here, the vector is (1,2,3)

plot_vectors([v], ['lightblue'])
plt.xlim(-1, 5)
plt.ylim(-1, 5)
```

# **Eigen Decomposition**

Eigen Decomposition of Matrix A would be

A = V . λ . V ^(-1)

Here, V is a matrix of all Eigen Vectors of Matrix A concatanated inside it

λ is a diagonal matrix, with the diagonal being Matrix A's EigenValues

```python
Mat_A = torch.tensor([[4,2],[-5,-3]], dtype=float)
Mat_A
```

```python
eigenval, eigenvec = torch.linalg.eig(Mat_A)
display(eigenval)
display(eigenvec)
```

```python
eigenvec
```

```python
eigenvecinv = torch.linalg.inv(eigenvec)
eigenvecinv
```

```python
Lambda = torch.diag(eigenval)
Lambda
```

proof that

A = V . λ . V ^(-1)

```python
Result_Matrix = torch.matmul(eigenvec,torch.matmul(Lambda, eigenvecinv))
Result_Matrix
```

# **Applications of SVD**

## Image compression and restoring

**Image Compression**

```python
from PIL import Image

monalisa = Image.open("/content/drive/MyDrive/monalisa.jpg") # for observing the change in rank use the franceflag.png in the drive
#for no change in rank, use monalisa.jpg
plt.imshow(monalisa)
```

```python
imggray = monalisa.convert('LA')
_ = plt.imshow(imggray)
```

```python
imgmat = np.array(list(imggray.getdata(band=0)), float)
imgmat.shape = (imggray.size[1], imggray.size[0])
imgmat = np.matrix(imgmat)
_ = plt.imshow(imgmat, cmap='gray')
```

```python
imgmat.shape
```

```python
Umonalisa, diagmonalisa, Vtransmonalisa = np.linalg.svd(imgmat)
```

Here, `diagonmonalisa` is an array. We would have to convert it into a diagonal matrix, add (674-445) more rows of 0s if you'd want to perform matrixmultiplication and restore the matrix.

which we are not going to do as we need to first compress the image.

currently, diagonmonalisa's shape is 455. mening there are 455 rows of values != 0 on the diagonal.

which means, (674-455) diagonals can be erased.

is that it? compression finished?

no. this is where we find something called the **Rank**

finding rank would remove even more dimensions which is not needed

## **Rank = largest dimension(row/column) x largest element in D x machine precision(e = 2.22e-16)**

```python
max_dim = np.max(imgmat.shape)
max_dim
```

```python
max_val = np.max(diagmonalisa)
max_val
```

```python
epsilon = 2.22e-16
epsilon
```

```python
tolerance = max_dim * max_val * epsilon
tolerance
```

```python
matrix_rank = np.sum(diagmonalisa > tolerance)
matrix_rank
```

```python
reconstimg = np.matrix(Umonalisa[:, :1]) * np.diag(diagmonalisa[:1]) * np.matrix(Vtransmonalisa[:1, :])
_ = plt.imshow(reconstimg, cmap='gray')
```

```python
for i in [2, 4, 8, 16, 32, 64]:
    reconstimg = np.matrix(Umonalisa[:, :i]) * np.diag(diagmonalisa[:i]) * np.matrix(Vtransmonalisa[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```

```python
reconstimg = np.matrix(Umonalisa[:, :matrix_rank]) * np.diag(diagmonalisa[:matrix_rank]) * np.matrix(Vtransmonalisa[:matrix_rank, :])
plt.imshow(reconstimg, cmap='gray')
title = "n = %s" % matrix_rank
plt.title(title)
plt.show()
```

# **Moore Penrose PseudoInverse**

We have learnt that taking inverse for a matrix is computationally cheaper than applying a transpose to it

Example : y = Xw

which when simplified for w

w = (X^-1).y

The Moore penrose inverse of matrix A can be represented as A^+

# **A^+ = U^T . D^+ . V**

```python
MatrixA = np.array([[-1,2],[3,-2],[5,7]])
MatrixA
```

```python
UMatrixA, DMatrixA, VtransMatrixA = np.linalg.svd(MatrixA)
```

```python
DmatrixAcomp = np.diag(DMatrixA)
DmatrixAcomp
```

```python
DMatrixAinv = np.linalg.inv(DmatrixAcomp)
DMatrixAinv
```

D^+ must have the same dimensions as A

```python
Dplus = np.concatenate ((DMatrixAinv, np.array ([[0, 0]]) .T) , axis=1)
Dplus
```

```python
Aplus = np.dot(VtransMatrixA.T, np.dot(Dplus, UMatrixA.T)) # we need V not Vtranspose and Utranspose not just U
Aplus
```

```python
np.linalg.pinv(MatrixA)
```

# **Trace Operator**
